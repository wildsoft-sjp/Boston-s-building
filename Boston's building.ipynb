## Influye la cercana al ro Charles en el valor de las viviendas? 

La alta gerencia desea saber si las viviendas cercanas al ro Charles tienen un valor significativamente diferente al resto. Para responder esta pregunta, se realiza una prueba de hiptesis utilizando el test t de Student para muestras independientes.


```python
from scipy import stats

# Separar los grupos segn cercana al ro
river_yes = df[df['chas'] == 1]['medv']
river_no = df[df['chas'] == 0]['medv']

# Prueba t de Student
t_stat, p_value = stats.ttest_ind(river_yes, river_no, equal_var=False)

# Resultados
print(f'Estadstico t: {t_stat:.2f}')
print(f'Valor p: {p_value:.4f}')

# Visualizacin
plt.figure(figsize=(8, 5))
sns.boxplot(x='chas', y='medv', data=df, palette='Set2')
plt.title('Valor medio de viviendas segn cercana al ro Charles')
plt.xlabel('Cercana al ro (0 = No, 1 = S)')
plt.ylabel('Valor medio (medv)')
plt.show()
```

## Conclusin 

El valor p obtenido es menor a 0.05, lo que indica que existe una diferencia estadsticamente significativa en el valor medio de las viviendas segn su cercana al ro Charles. Las viviendas cercanas al ro tienden a tener un valor ms alto, lo que sugiere que la ubicacin geogrfica influye en el precio del mercado inmobiliario.



## Influye la antigedad de las viviendas en su valor? 

La gerencia desea saber si las viviendas construidas antes de 1940 tienen un valor significativamente diferente al resto. Para ello, se crea una variable categrica (`old_house`) y se aplica una prueba t de Student para comparar los grupos.


```python
# Crear variable categrica: viviendas construidas antes de 1940
df['old_house'] = df['age'].apply(lambda x: 1 if x > 60 else 0)

# Separar los grupos
old = df[df['old_house'] == 1]['medv']
new = df[df['old_house'] == 0]['medv']

# Prueba t de Student
t_stat, p_value = stats.ttest_ind(old, new, equal_var=False)

# Resultados
print(f'Estadstico t: {t_stat:.2f}')
print(f'Valor p: {p_value:.4f}')

# Visualizacin
plt.figure(figsize=(8, 5))
sns.boxplot(x='old_house', y='medv', data=df, palette='Set3')
plt.title('Valor medio segn antigedad de la vivienda')
plt.xlabel('Antigedad (0 = 1940, 1 = >1940)')
plt.ylabel('Valor medio (medv)')
plt.show()
```

## Conclusin 

El valor p obtenido indica si hay diferencia significativa entre viviendas antiguas y nuevas. Si el valor p < 0.05, se concluye que las viviendas construidas antes de 1940 tienen un valor diferente al resto. En este caso, se observa que las viviendas ms nuevas tienden a tener un valor medio ms alto, lo que puede reflejar mejoras en infraestructura, ubicacin o demanda del mercado.
